```python
def categorize_aqi(aqi):
    if aqi <= 50:
        return 1
    elif aqi <= 100:
        return 2
    elif aqi <= 200:
        return 3
    elif aqi <= 300:
        return 4
    else:
        return 5

# Apply the function to create the 'range' column
delhi_df['range'] = delhi_df['AQI'].apply(categorize_aqi)
delhi_df
```
