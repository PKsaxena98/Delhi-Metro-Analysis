#Zomato maximun transaction/order mainly happen in India US UK

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0 : 'Rating Count'})
```

```python
ratings
```

## Observation 
1 . When the rating is b/w 4.5 to 4.9 is Excellent
2 . When the rating is b/w 4.0 to 4.4 is Very Good
3 . When the rating is b/w 3.5 to 3.9 is Good
4 . When the rating is b/w 3.0 to 3.4 is average
5 . When the rating is b/w 2.5 to 2.9 is average
6 . When the rating is b/w 2.0 to 2.4 is poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,5)
sns.barplot(x = "Aggregate rating",  y = "Rating Count", hue='Rating color',data=ratings , palette = ['blue','red' ,'Orange' ,'yellow' , 'green', 'green'])
```

Observation
1 . Not rated count is very high
2 . Maximum number of rating are is average (2.5to3.9)

sns.countplot(x="Rating color" ,  data=ratings ,palette = ['blue','red' ,'Orange' ,'yellow' , 'green', 'green'] )
